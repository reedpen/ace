This tutorial demonstrates how to align independent miniscope and electrophysiology recorders using synchronizing TTL pulses.

## 1. Setup

```python
import ace_neuro
from ace_neuro.pipelines.multimodal import MultimodalPipeline
from pathlib import Path
import matplotlib.pyplot as plt

print(f"Using ACE-neuro version: {ace_neuro.__version__}")
```

## 2. The Alignment Logic

Multimodal analysis in ACE-neuro relies on finding the exact correspondence between calcium imaging frames and ephys sampling indices. This is achieved by:
1. Identifying TTL pulses in both streams representing common events (e.g., experiment start/end).
2. Generating a linear transformation to map timestamps between recorders.
3. Correcting for dropped frames or recording drift.

```python
project_path = Path("path/to/your/project")
data_path = Path("path/to/your/raw_data")
line_number = 97

pipeline = MultimodalPipeline()

pipeline.run(
    line_num=line_number,
    project_path=project_path,
    data_path=data_path,
    channel_name="PFCLFP",
    ca_events=True  # Enable calcium event alignment
)
```
