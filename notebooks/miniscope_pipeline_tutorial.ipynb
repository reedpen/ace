# ACE-neuro: Miniscope Pipeline Tutorial

This notebook demonstrates the end-to-end workflow for processing 1-photon calcium imaging data using the ACE-neuro `MiniscopePipeline`.

## 1. Environment Setup

Ensure you have ACE-neuro installed in your `caiman` environment.

```python
import ace_neuro
from ace_neuro.pipelines.miniscope import MiniscopePipeline
from pathlib import Path
import matplotlib.pyplot as plt

print(f"Using ACE-neuro version: {ace_neuro.__version__}")
```

## 2. Configuration

ACE-neuro uses a CSV-driven approach for experiment management. You must provide a `project_path` containing `experiments.csv` and `analysis_parameters.csv`.

```python
# Define your paths explicitly
project_path = Path("path/to/your/project")
data_path = Path("path/to/your/raw_data")
line_number = 96  # Matches the 'line number' in your CSVs
```

## 3. Running the Pipeline

The `MiniscopePipeline` orchestrates:
1. **Preprocessing**: Cropping and detrending.
2. **Processing**: Motion correction and CNMF-E source extraction.
3. **Post-processing**: Component evaluation and event detection.

### Interactive vs. Headless Mode
- Set `headless=False` (default) to use interactive GUIs for cropping and neuron curation.
- Set `headless=True` for automated batch processing on clusters.

```python
pipeline = MiniscopePipeline()

# Run the pipeline
# Note: This will open GUIs if headless=False
pipeline.run(
    line_num=line_number,
    project_path=project_path,
    data_path=data_path,
    headless=True 
)
```

## 4. Visualizing Results

After processing, you can access the results directly from the `MiniscopeDataManager`.

```python
dm = pipeline.miniscope_data_manager

print(f"Detected {len(dm.ca_events_idx)} neurons.")

# Example: Plot the spatial footprints of detected neurons
plt.figure(figsize=(10, 10))
plt.imshow(dm.Cn, cmap='gray')
plt.title("Max Projection with Spatial Footprints")
plt.show()
```
