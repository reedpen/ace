# ACE-neuro: Ephys Pipeline Tutorial

This notebook demonstrates how to process multi-channel electrophysiology data using the ACE-neuro `EphysPipeline`.

## 1. Environment Setup

```python
import ace_neuro
from ace_neuro.pipelines.ephys import EphysPipeline
from pathlib import Path
import matplotlib.pyplot as plt

print(f"Using ACE-neuro version: {ace_neuro.__version__}")
```

## 2. Configuration

The ephys pipeline requires Neuralynx (`.ncs`) or ONIX format data. Metadata is pulled from `experiments.csv` based on the specified experiment line number.

```python
project_path = Path("path/to/your/project")
data_path = Path("path/to/your/raw_data")
line_number = 96
channel_name = "PFCLFP" # Example channel
```

## 3. Running the Pipeline

The `EphysPipeline` handles:
- Raw data ingestion
- Artifact identification and removal
- Bandpass filtering
- Spectral analysis (Spectrograms and Power Spectral Density)

```python
pipeline = EphysPipeline()

pipeline.run(
    line_num=line_number,
    project_path=project_path,
    data_path=data_path,
    channel_name=channel_name,
    remove_artifacts=True,
    filter_range=[0.5, 4],  # Delta band
    plot_spectrogram=True
)
```

## 4. Signal Inspection

You can access the processed `Channel` objects and their associated metadata directly.

```python
channel = pipeline.ephys_data_manager.get_channel(channel_name)

plt.figure(figsize=(15, 5))
plt.plot(channel.time_vector, channel.signal_filtered)
plt.title(f"Filtered Signal: {channel_name}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0, 30)  # View first 30 seconds
plt.show()
```
